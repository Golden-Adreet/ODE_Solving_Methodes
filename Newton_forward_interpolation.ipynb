```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
x=[5,10,15,20,25,30]
y=[45,105,174,259,364,496]
#interpolation using scipy
y2=interp1d(x,y)
#default is linear
y3=interp1d(x,y,kind='cubic')
print (y2)
plt.plot(x,y,'o',x,y2(x),x,y3(x),'-')
plt.show()
dt = [y,]
for i in range(len(x)-1):
  f = []
  for j in range(len(x)-i-1):
    a = dt[i][j+1] - dt[i][j]
    f.append(a)
  dt.append(f)
def cal(u,n):
  a = 1
  for i in range(n):
    a*=(u-i)
  return a
def fact(n):
  if n == 0:
    return 1
  else:
    return n*fact(n-1)
val = float(input('value to interpolate:'))
u = (val-x[0])/(x[1]-x[0])
iv = dt[0][0]
for i in range(1,len(x)):
  iv+=(cal(u,i)/fact(i))*dt[i][0]
print(f'the interpolated value is given by:{iv}')
```
